# What this script does

In this script:

1. Reduce the CMS data to include only **deficiencies** whose F-tags:
    - belong to the "Infection Control" tag group *and*
    - have an equivalent WAC code
1. Reduce the scraped data to include only **penalties** whose WAC code:
    - belongs to the "Infection Control" section

# I. SETTINGS

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

# II. DATA

### II.1. Import

```python
# Deficiencies (from CMS database)
df_sod_wa_orig = pd.read_csv('../C_output_data/sod_wa.csv', 
                             dtype='object', parse_dates=['inspection_dt'])

# Penalties (scraped from the enforcement letter PDFs)
df_letters_orig = pd.read_csv('../C_output_data/scraped_data_v2.csv', 
                              parse_dates=['letter_dt', 'survey_dt'])

# Facilities with COVID cases
df_covid_orig = pd.read_csv('../C_output_data/homes_with_covid_20200708.csv')

# WAC code details
df_wac = pd.read_csv('../C_output_data/wac_codes_df_t338c97.csv')
```

### II.2. Create working copies

```python
# Deficiencies (from CMS database)
df_sod_wa = df_sod_wa_orig.copy()

# Penalties (scraped from the enforcement letter PDFs)
df_letters = df_letters_orig.copy()

# Facilities with COVID cases
df_covid = df_covid_orig.copy()
```

### II.3. Reduce the data to only infection control-related cases

#### II.3.1. Deficiency data

Select only tags whose group name is "Infection Control"

```python
df_ic_deficiencies = df_sod_wa[df_sod_wa['tag_group_name']=='Infection Control']
df_ic_deficiencies = pd.DataFrame(df_ic_deficiencies['tag'].value_counts(dropna=False))
df_ic_deficiencies.columns = ['deficiency_count']

df_ic_deficiencies
```

Let's add the tag titles, which describe what specific subject inside infection control the tag is about.

```python
temp = [['0441', 'Infection Prevention & Control'],
        ['0880', 'Infection Prevention & Control'],
        ['0334', 'Influenza and Pneumococcal Immunizations'],
        ['0883', 'Influenza and Pneumococcal Immunizations'],
        ['0881', 'Antibiotic Stewardship Program']]
temp = pd.DataFrame(temp, columns=['tag', 'tag_title'])

df_ic_deficiencies = df_ic_deficiencies.join(temp.set_index('tag'), how='left').reset_index()
df_ic_deficiencies.columns = ['tag', 'deficiency_count', 'tag_title']
del(temp)

df_ic_deficiencies
```

Now we select only IC-related tags to those that have a corresponding WAC code

All of the above F-tags are belong to the "Infection Control" tag group. But does that mean we should consider all of them when counting the number of defficiencies recorded for infection control?

**F883** and **F334** are the same title, *Influenza and Pneumococcal Immunizations* (F883 is the updated version of F334.) There is a state code—[WAC 388-97-1340](https://app.leg.wa.gov/WAC/default.aspx?cite=388-97-1340)—whose section (the equivalent of the F-tag group) has the same classification: *Influenza and pneumococcal immunizations*. So we should include F883 & F334 in our defficiency counts.

**F881**, Antibiotic Stewardship Program, there is no equivalent WAC code. That means DSHS can't fine a facility for it. There is probably something to be said about such federal deficiency not being contemplated in the local regulations.
Why doesn’t the state review such a requirement, when the federal gov does? But it is not in the scope of our reporting, so for now we exclude these 83 deficiencies from our count.

```python
df_ic_deficiencies = df_ic_deficiencies[df_ic_deficiencies['tag']!='0881'].reset_index(drop=True)

df_ic_deficiencies
```

Now that we have the list of tags that represent infection control deficiencies, reduce *df_sod_wa* to include only those tags.

```python
df_sod_wa_ic = df_sod_wa[df_sod_wa['tag'].isin(df_ic_deficiencies['tag'])]
df_sod_wa_ic = df_sod_wa_ic.reset_index(drop=True)

# Consistency test
assert len(df_sod_wa_ic) == df_ic_deficiencies['deficiency_count'].sum()

print(df_sod_wa_ic.shape)
df_sod_wa_ic.nunique()
```

#### II.3.2. Penalization data (scraped data)

Get all the WAC codes related to infection control

```python
df_ic_wacs = df_wac[df_wac['section'] == 'Infection Control'].reset_index(drop=True)
df_ic_wacs = df_ic_wacs.drop(['section', 'sub_chp_num', 'sub_chp_name'], axis=1)
df_ic_wacs.columns = ['wac', 'wac_desc']

df_ic_wacs
```

Reduce the scraped data to only the records that show IC-related WACs

```python
df_letters_ic = df_letters[df_letters['wac_short'].isin(df_ic_wacs['wac'])]
df_letters_ic = df_letters_ic.reset_index(drop=True)
```

After all this filtering, we now have the enforcement letters that issued IC-related CMPs, wich we can see here:

```python
print(df_letters_ic['pdf_name'])

# for pdf in df_letters_ic['pdf_name']:
#     print('open', pdf.replace(' ', '\ ').replace('(', '\(').replace(')', '\)'))
```

A visual inspection of these letters revealed that one of the two for Linden Grove is just an updated version of the other. So we eliminte the oritinal one.

```python
# Eliminate the original letter to Linen Grove
df_letters_ic = df_letters_ic[df_letters_ic['pdf_name']!='Linden Grove (IJ, R, Cont sp, Cont Cond, CF) 5 21 20.pdf']
df_letters_ic = df_letters_ic.reset_index(drop=True)
```

```python
print(df_letters_ic['cmp_item'].sum())
df_letters_ic
```

#### Total-CMP vs IC-CMP

```python
temp = df_sod_wa_ic['facility_id'].value_counts(dropna=False)
temp = pd.DataFrame(temp)
temp.columns = ['num_defs']
temp
```

#### Add the total IC-related CMPs per facility

```python
temp2 = df_letters_ic.groupby(['manual_cms_number'])[['cmp_item']].sum()
temp2 = temp2.reset_index().set_index('manual_cms_number')
print(temp2['cmp_item'].sum())
temp2
```

```python
temp3 = df_sod_wa[['facility_id', 'facility_name']]
temp3 = temp3.drop_duplicates()

# Remove duplicates
print(temp3['facility_id'].value_counts().head())
print('\n')
print(temp3[temp3['facility_id'].isin(['505017','505254'])].sort_values('facility_id'))

temp3 = temp3[~temp3['facility_id'].duplicated()]
temp3 = temp3.set_index('facility_id')
temp3
```

```python
temp4 = df_sod_wa_ic[['facility_id', 'severity_code']]
temp4 = temp4[temp4['severity_code'].isin(['G','J','K'])]
# temp4 = temp4.set_index('facility_id')
temp4
```

```python
df_facs_ic = temp.join(temp2, how='outer').join(temp3, how='outer')
del(temp, temp2, temp3)

df_facs_ic = df_facs_ic.reset_index()
df_facs_ic.columns = ['manual_cms_number','num_defs','cmp_item', 'facility_name']
df_facs_ic = df_facs_ic[['facility_name', 'manual_cms_number', 'num_defs', 'cmp_item']]
df_facs_ic = df_facs_ic.sort_values(['num_defs'], ascending=False)
df_facs_ic = df_facs_ic.reset_index(drop=True)

# Add a boolean column regarding the presence of covid cases
df_facs_ic['covid_cases'] = df_facs_ic['manual_cms_number'].isin(df_covid['cms_num'])
# Add a boolean column that indicates if that facility had top severe deficiencies
df_facs_ic['top_severity'] = df_facs_ic['manual_cms_number'].isin(temp4['facility_id'])

# Consistency tests:

# Number of faciliites that registered IC deficiencies
assert len(df_sod_wa_ic) == df_facs_ic['num_defs'].sum()
print(len(df_sod_wa_ic))
# Number of deficiencies
assert df_sod_wa_ic['facility_id'].nunique() == df_facs_ic[~df_facs_ic['num_defs'].isna()]['manual_cms_number'].nunique()
print(df_sod_wa_ic['facility_id'].nunique())
# Total amount of IC fines
assert df_letters_ic['cmp_item'].sum() ==  df_facs_ic['cmp_item'].sum()
print(df_letters_ic['cmp_item'].sum())

# # Hard to compare here, so we export to a CSV
# df_facs_ic.to_csv('/Users/mvilla/Downloads/ic_TRASHABLE.csv', index=False)
```

```python
df_facs_ic
```

```python
num_facs_defs = (~df_facs_ic['num_defs'].isna()).sum()
num_facs_covid = df_facs_ic['covid_cases'].sum()
num_facs_both = ((~df_facs_ic['num_defs'].isna()) & df_facs_ic['covid_cases']).sum()
num_facs_covid_no_defs = ((df_facs_ic['num_defs'].isna()) & df_facs_ic['covid_cases']).sum()

print('There were', num_facs_defs, 'facilities that had IC deficiencies.')
print('There were', num_facs_covid, 'facilities that had COVID cases.')
print('There were', num_facs_both, 'facilities that had both.')
print('\r')

print('So if a facility registered an IC deficiency, there is a', 
      round(num_facs_both/num_facs_defs*100), '% chance it also registered COVID cases.')
print('But if a facility registered a COVID case, there is a', 
      round(num_facs_both/num_facs_covid*100), '% chance it had IC deficiencies in the past.')
print('Only', num_facs_covid_no_defs, 'of the', num_facs_covid, 'facilities with covid cases had no deficiency history.')
```
