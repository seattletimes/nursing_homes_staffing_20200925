# Settings

```python
import pdfplumber
import pandas as pd
import re
from decimal import Decimal
import numpy as np
```

# Explanatory note

The **Inspection** reports of nursing homes are structured around a large table in each page, as can be seen in the sample page below.  Some of the information we want is contained inside that table and some outside of it. The algorithms required to scrap the information the inside and outside of the table are very different, so we decided to write them in separate notebooks.

This script scraps the information contained **outside** the table.

```python
sample_pdf = '/Volumes/files/COVID19/DSHS_Facility_Reports_Manuel/DSHS_ALTSA_Facility_Reports/reports_pdfs/NH_3333/R  St. Francis of Bellingham Survey QMTV11 12-4-18 - KM.pdf'
sample_pdf = pdfplumber.open(sample_pdf)
sample_pdf.pages[1].to_image(resolution=150)
```

# Global variables

#### Path to all the reports

```python
# This is where **all** the nursing facility PDFs are, whether they are inspection reports or others.
path = '/Volumes/files/COVID19/DSHS_Facility_Reports_Manuel/DSHS_ALTSA_Facility_Reports/reports_pdfs/NH_3333/'
```

#### Create a list of all the unique pdf_names of the *inspection* reports for nursing home facilities

```python
# Read the CSV that contains the metadata for all nursing home reports
df_reports = pd.read_csv('../C_output_data/reports_metadata/reports_metadata_nhf.csv')

# Reduce df_reports to **Inspection** reports only
inspection_reports = df_reports[df_reports['rep_type']=='Inspection']

# Get a list of all the unique **Inspection** reports' pdf_names
inspection_reports = inspection_reports['pdf_name'].unique()
inspection_reports = pd.Series(inspection_reports)
inspection_reports = inspection_reports.sort_values().reset_index(drop=True)

print('There are', len(inspection_reports), 'inspection reports of nursing home facilities')
```

#### Create the dataframe where all the data will be deposited

```python
df = pd.DataFrame(columns = ['pdf_name', 'num_pages', 'page', 'num_tables', 
                             'omb_no', 'provider_id', 'survey_comp_date', 
                             'provider_name', 'provider_address',
                             'form_id', 'event_id', 'facility_id'])
```

# Scraping fest

```python
# Each iteration of this loop represnts one PDF report
for index, pdf_name in inspection_reports.iteritems():

    pdf = pdfplumber.open(path + pdf_name)
    
    # Each iteration of this loop represnts one page inside a PDF report
    for pg in pdf.pages:
        
        # In some pages, no tables are detected. It could be there are indeed none, 
        # or pdfplumber couldn't find them. Regardless the reason, this breaks the algorithm, so
        # if we don't find any tables in a particular page, we skip the whole page.
        if not pg.find_tables():
            new_record ={'pdf_name': pdf_name,
                         'num_pages':len(pdf.pages),
                         'page':pg.page_number,
                         'num_tables':len(pg.find_tables()), # Since no tables were find, this is = 0
                         'omb_no':np.nan,
                         'provider_id':np.nan,
                         'survey_comp_date':np.nan,
                         'provider_name':np.nan,
                         'provider_address':np.nan,
                         'form_id':np.nan,
                         'event_id':np.nan,
                         'facility_id':np.nan}
        
        # If tables are indeed detected in the page:
        else:                
            # After observing a large sample of reports, we realized that most pages are structured into one 
            # big table that is divided in several sections. So the pg.find_tables() method will produce a list
            # of only one element (one table) for each page.
            tb = pg.find_tables()[0]
            
            
            # Top area: omb number
            # This data point lies outside the table, so we need to delimit the area where we can find it.
            top_area = pg.crop((Decimal('0'), Decimal('0'),  
                                tb.bbox[2], tb.bbox[1]))            
            omb_no = top_area.extract_text().split(' ')[-1]

            
            # Row 0: provider ID and survey completion date
            cells = [c for c in tb.rows[0].cells if c]
            # Provider ID
            provider_id = pg.crop(cells[1]).extract_text().split('\n')[-1].strip()
            # Survey completion date
            survey_comp_date = pg.crop(cells[3]).extract_text().split('\n')[-1].strip()


            # Row 1: Provider name and address
            del(cells)
            cells = [c for c in tb.rows[1].cells if c]
            # In some pages, row 1 lacks a dividing line between the name and address areas
            # (Again, we realized that after a lot of observations.)
            # Those cases break the algorithm, so for pages with this problem, we skip these two data items.
            if len(cells) == 1:
                provider_name = np.nan
                provider_address = np.nan                
            else:
                provider_name = re.sub('^[^\n]+', '', pg.crop(cells[0]).extract_text()).strip()
                provider_address = re.sub('^[^\n]+', '', pg.crop(cells[1]).extract_text()).strip()
                provider_address = re.sub('\n', ', ', provider_address)
                provider_address = re.sub('\s{2,}', ' ', provider_address).strip()


            # Bottom area: form, event and facility IDs
            # These data points lie outside the table, so we need to delimit the area where we can find them.
            bottom_area = pg.crop((tb.bbox[0],
                                   tb.bbox[3], 
                                   pg.bbox[2], 
                                   pg.bbox[3]))
            # On some pages, the text in the bottom area is too convoluted and breaks the algorithm.
            # For those pages, we skip these data points.
            try:    form_id = re.search('FORM\s([^\s]+)\s.*$', bottom_area.extract_text()).group(1)
            except: form_id = np.nan

            try:    event_id = re.search('Event\sID:\s?([^\s]+)\s.*$', bottom_area.extract_text()).group(1)
            except: event_id = np.nan

            try:    facility_id = re.search('Facility\sID:\s?([^\s]+)\s.*$', bottom_area.extract_text()).group(1)
            except: facility_id = np.nan

            # Organize all the extracted data items into a single record:
            new_record ={'pdf_name': pdf_name,
                         'num_pages':len(pdf.pages),
                         'page':pg.page_number,
                         'num_tables':len(pg.find_tables()),
                         'omb_no':omb_no,
                         'provider_id':provider_id,
                         'survey_comp_date':survey_comp_date,
                         'provider_name':provider_name,
                         'provider_address':provider_address,
                         'form_id':form_id,
                         'event_id':event_id,
                         'facility_id':facility_id}
        
        # Append the new record in the dataframe
        df = df.append(new_record, ignore_index=True)
        del(new_record)

    print (index, '| Pages:', len(pdf.pages), '|', pdf_name )
```
