# III.1. Types of enforcement taken

```python
print(len(df_cmp))

df_actions = df_cmp[['pdf_name', 'letter_dt', 'action']]
# df_actions = df_actions[~df_actions['action'].isna()]
df_actions = df_actions.drop_duplicates()
print(len(df_actions))

# Confirm that each record corresponds to an unique pdf_name.
# (This will also confirm that each pdf_name has only one action.)
assert len(df_actions) == df_actions['pdf_name'].nunique()

# Reduce
df_actions = df_actions.dropna(axis=0, subset=['action'])
df_actions = df_actions.reset_index(drop=True)
print(len(df_actions))

# After a visual analysis of df_cmp['action'].value_counts(),
# we come up with the following four types of actions that the DSHS seems to have taken:
df_actions['civil_fine'] = df_actions['action'].str.lower().str.contains('civil fine')
df_actions['stop_placement'] = df_actions['action'].str.lower().str.contains('stop placement')
df_actions['correction_plan'] = df_actions['action'].str.lower().str.contains('plan of correction')
df_actions['conditions_on_lic'] = df_actions['action'].str.lower().str.contains('conditions')
df_actions = df_actions.drop(['pdf_name', 'action'], axis=1)
```

```python
df_actions
```

**IMPORTANT**: A very few (roughly a dozen) of the actions were reversals. That is, they lifted (instead of imposing) a stop placement, or condions on a license, or a fine. To view, use ***df_cmp['action'].value_counts()***

Now let's see if we can trust that the 'action' column indeed captures all actions taken. We test it out with CMPs, which is the one action for which we already have its own column, *cmp_agg*.

```python
print('Number of letters where we found an action section:',
df_cmp[~df_cmp['action'].isna()]['pdf_name'].nunique())

print('Number of letters with action section that included a civil fine:',
df_actions['civil_fine'].sum())

print('Number of letters with a non-NaN value for cmp_agg:',
df_cmp[~df_cmp['cmp_agg'].isna()]['pdf_name'].nunique())
```

**IMPORTANT**: So not all letters that imposed a civil fine had a section in the front page that described the actions to be taken (if any), which would include such civil fine. In other words, the 'actions' column does not capture all actions taken.

If, however, we assume that the those blanks are not significative, the data suggests that the DSHS relied mostly on CMPs and very little in other types of enforcements:

```python
print(df_actions.sum())

piv = df_actions.pivot_table(index='letter_dt', 
                             aggfunc='sum', 
                             fill_value = 0)
piv = piv.resample('M').sum()
piv.plot(title='Number of Enforcements - By Type')
```

```python
print('The enforcement method of choice was by large civil monetary penalties.',
      'According to our analysis, out of the', 
      df_cmp['pdf_name'].nunique(), 
      'enforcement letters reviewed,', 
      df_cmp[~df_cmp['cmp_agg'].isna()]['pdf_name'].nunique(),
      'of them issued such fines.\n')

print('Stop placements, the next most used enforcement method, was only found to be issued',
      'in', df_actions['stop_placement'].sum(), 'letters.',
      'The imposition of conditions on a facility s license was used', df_actions['conditions_on_lic'].sum(),
      'times, while the establishment of correction plans was only issued',
      df_actions['correction_plan'].sum(), 'times.')
```

# III.2. CMP

## III.2.1. CMP Summary

```python
df_enf_summary = df_cmp[~df_cmp['cmp_item'].isna()]

print('During the period in question, a total of $', 
      round(df_cmp['cmp_agg'].sum()/1e6,1), 'MM were imposed as CMPs to',
      df_enf_summary['manual_cms_number'].nunique(), 'facilities. This represents an average fine of $',
      round(df_cmp['cmp_agg'].sum()/df_enf_summary['manual_cms_number'].nunique()), 'per facility.')
```

```python
# Compute the number of penalties issued per section
df_summary = pd.DataFrame(df_cmp['section'].value_counts())
df_summary.columns = ['num_penalties']

# Compute the amount of CMP fined per section
df_enf_summary = df_cmp[['section', 'cmp_item']].groupby('section').sum()

# Compute the % contribution of total CMP per section
# NOTE: This column will not add up to 100% because of the 'cmp_item' cases
# that the scraper could not scrap
df_enf_summary['cmp_pct'] = df_enf_summary['cmp_item']/df_cmp['cmp_agg'].sum()
df_summary = df_summary.join(df_enf_summary, how='outer')

# Compute the average penalty per section
df_summary['cmp_avg'] = df_summary['cmp_item']/df_summary['num_penalties']

# Compute the number of deficiencies per section
df_enf_summary = df_sod_wa['tag_group_name'].value_counts()
df_enf_summary = pd.DataFrame(df_enf_summary)
df_enf_summary.columns = ['num_deficiencies']
df_summary = df_summary.join(df_enf_summary, how='outer')

# Compute the penalization rate
df_summary['penalization_rate'] = df_summary['num_penalties']/df_summary['num_deficiencies']


df_summary = df_summary.sort_values('cmp_item', ascending=False)
```

```python
df_summary
```

```python
piv = df_cmp.pivot_table(index='letter_dt', 
                             values='cmp_agg', 
                             aggfunc='sum', 
                             fill_value = 0)
piv = piv.resample('M').sum()

print(piv['cmp_agg'].sum())
print(df_cmp['cmp_agg'].sum())

piv.plot(title='Civil Monetary Penalties 2017-2020 (Total = $' + str(round(piv['cmp_agg'].sum()/1e6,1)) + 'MM)')
```

## III.2.2. CMPs by section

```python
df_enf_summary = df_cmp[df_cmp['section'].isin(['Nursing Services'])]
cmp_ns = df_enf_summary['cmp_item'].sum()

df_enf_summary = df_cmp[df_cmp['section'].isin(['Quality of Care'])]
cmp_qc = df_enf_summary['cmp_item'].sum()

df_enf_summary = df_cmp[df_cmp['section'].isin(['Infection Control'])]
cmp_ic = df_enf_summary['cmp_item'].sum()

print('Most of the fines â€” $', round((cmp_ns+cmp_qc)/1e6,1), 'MM, or', round((cmp_ns+cmp_qc)/df_cmp['cmp_agg'].sum()*100),
      '% of the total â€” were triggered for deficiencies in two WAC sections:\n')

print('Nursing services â€” $', round(cmp_ns/1e3), 'K\n',
      'Quality of care â€” $', round(cmp_qc/1e3), 'K\n')

print('In contrast, infection control fines amounted to a mere $', round(cmp_ic/1e3,1),'K , just',
      round(cmp_ic/df_cmp['cmp_agg'].sum()*100,1), '% of the total.')
```

# III.3. Enforcements by Facility - Summary

```python
df_non_cmp = df_cmp.copy()

# Create one boolean column per each non-cmp action type
df_non_cmp['stop_placement'] = df_non_cmp['action'].str.lower().str.contains('stop placement')
df_non_cmp['correction_plan'] = df_non_cmp['action'].str.lower().str.contains('plan of correction')
df_non_cmp['conditions_on_lic'] = df_non_cmp['action'].str.lower().str.contains('conditions')


# For each type of non-CMP enforcement, make a list of the facilities that were receivers of such enforcements.
df_stop = df_non_cmp[df_non_cmp['stop_placement']==True]
df_stop = df_stop[['manual_cms_number', 'facility_name']]
df_stop = df_stop.drop_duplicates().reset_index(drop=True)

df_plan = df_non_cmp[df_non_cmp['correction_plan']==True]
df_plan = df_plan[['manual_cms_number', 'facility_name']]
df_plan = df_plan.drop_duplicates().reset_index(drop=True)

df_cond = df_non_cmp[df_non_cmp['conditions_on_lic']==True]
df_cond = df_cond[['manual_cms_number', 'facility_name']]
df_cond = df_cond.drop_duplicates().reset_index(drop=True)

# Consolidate into a summary of enforcements
df_enf_summary = df_cmp[['manual_cms_number', 'facility_name', 'cmp_agg']]
df_enf_summary = df_enf_summary.groupby(['manual_cms_number', 'facility_name']).sum().sort_values('cmp_agg', ascending=False)
df_enf_summary = df_enf_summary.reset_index()

df_enf_summary['stop'] = df_enf_summary['manual_cms_number'].isin(df_stop['manual_cms_number'])
df_enf_summary['plan'] = df_enf_summary['manual_cms_number'].isin(df_plan['manual_cms_number'])
df_enf_summary['condition'] = df_enf_summary['manual_cms_number'].isin(df_cond['manual_cms_number'])

df_enf_summary.to_csv('/Users/mvilla/Downloads/enforcements_by_facility_summary.csv', index=False)
df_enf_summary
```
