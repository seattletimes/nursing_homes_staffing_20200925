# Settings

```python
import pandas as pd
import pdfplumber
import numpy as np
from os import listdir
import re
```

Now that we have downloaded the 4.6+ **adult family reports**, we want to know if they are text based or image based. We will go through each page of all the reports and confirm if the page is image-based or text-based.

**Note**: This step could have been naturally implemented in script **Step1_Downloading_Reports_AF**, but I didn't think of it then. To implement and run there now we would need to re-download all the reports—not a very sane choice—so carrying out this task in a separate script instead.

```python
# Obtain a list of all the downloaded adult family PDF reports
path = '/Volumes/files/COVID19/DSHS_Facility_Reports_Manuel/reports_pdfs/AF_4644/'
file_list = listdir(path)

# Weed out any files in the folder that are not PDFs
file_list = [file for file in file_list if re.search('\.pdf$', file)]
file_list = pd.Series(file_list)
```

```python
print('There is a total of', len(file_list), 'reports for adult family facilities.')
print('We wish to know which of those reports contain image-based pages.')
```

# A very long loop...

```python
df = pd.DataFrame(columns = ['pdf_name', 'page', 'image_based'])

for index, pdf_name in file_list.iteritems():
    
    print(index, '|', pdf_name)
    
    try:
        pdf = pdfplumber.open(path + pdf_name)
        for pg in pdf.pages:
            new_record = {'pdf_name':pdf_name,
                          'page':pg.page_number,
                          'image_based':not pg.extract_text()}
            df = df.append(new_record, ignore_index=True)

    # If the PDF is damaged or othewise not readable
    except:
        new_record = {'pdf_name':pdf_name,
                      'page':np.nan,
                      'image_based':np.nan}
        df = df.append(new_record, ignore_index=True)
```

# Saving results

```python
df.to_csv('../C_output_data/reports_metadata/image_based_reports_af.csv', index=False)
# df = pd.read_csv('../output_data/reports_metadata/image_based_reports_af.csv')
```

### How many *reports* have at least one page that is  image-based?

```python
temp = df.copy()
temp = temp[~temp['image_based'].isna()]
temp = temp[temp['image_based']]

# Save down this number for consistency test later
num_image_pdfs = temp['pdf_name'].nunique()

print('So', num_image_pdfs, 'out of the total', df['pdf_name'].nunique(),'adult family reports contain at least one image-based page.')
print('That is,', round(num_image_pdfs / df['pdf_name'].nunique() * 100),'percent.')
```

### Breakdown of image-based reports by report type

How many image-based reports we had inside each report category?

```python
del(temp)
temp = df[~df['image_based'].isna()]
temp = temp[temp['image_based']]
temp.drop(columns=['page', 'image_based'], inplace=True)
temp.drop_duplicates(inplace=True)
temp.reset_index(drop=True, inplace=True)
temp = temp.groupby('rep_type').count().reset_index()
temp.columns = ['rep_type', 'image_based']

assert temp['image_based'].sum() == num_image_pdfs
```

```python
temp
```

Now let's see what proportion of reports in each category were image-based.

```python
temp2 = df.copy()
temp2 = temp2[['pdf_name', 'rep_type']]
temp2 = temp2.drop_duplicates()
temp2 = temp2.groupby('rep_type').count().reset_index()
```

```python
temp2
```

```python
# Join
temp = temp.join(temp2.set_index('rep_type'), on='rep_type', how='outer')

# Reograize and rename columns
temp = temp[['rep_type', 'pdf_name', 'image_based']]
temp.columns = ['rep_type', 'total', 'image_based_total']

# Compute the proporion of image-based reports in each category
temp['pct_image_based'] = temp['image_based_total']/temp['total']

# Consisntency tests
assert temp['total'].sum() == len(file_list)
assert temp['image_based_total'].sum() == num_image_pdfs
```
